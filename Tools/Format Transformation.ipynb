```python
import os
import rawpy
import imageio
```

```python
def convert_arw_to_jpg(arw_file, jpg_file):
    with rawpy.imread(arw_file) as raw:
        rgb = raw.postprocess()
    imageio.imwrite(jpg_file, rgb)

# Replace 'input.arw' with the path to your ARW file
# Replace 'output.jpg' with the path where you want to save the JPEG file
convert_arw_to_jpg('/Users/simonleung/Desktop/王然师姐0425图片/_DSC8176.ARW', '/Users/simonleung/Desktop/王然师姐0425图片/_DSC8176.jpg')


```

```python
folder = os.listdir('/Users/simonleung/Desktop/王然师姐0425图片/')
output_folder = '/Users/simonleung/Desktop/王然师姐0425图片/jpg/'
i = 0
for file in folder:
    f, e = os.path.splitext(file)
    if e == '.ARW':
        input_file = '/Users/simonleung/Desktop/王然师姐0425图片/' + file
        output_file = output_folder + f + '.jpg'
        convert_arw_to_jpg(input_file, output_file)
        i += 1
    print(file + ' ' + str(i) + '/' + '212')
```
